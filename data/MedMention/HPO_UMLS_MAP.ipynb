```python
import pandas as pd
hpo_umls_df=[]
count=0 # 15332
with open('hpo-umls.txt','r') as file_read:
    main_list=[]
    while count<15332:
        each_line=file_read.readline()
        if(each_line.find('[Term]')>=0):
            count+=1
            each_line=file_read.readline()
            each_hpo_umls=[]
            if(each_line.find('id: HP:')>=0):
                each_hpo_umls.append(each_line.strip().split(':')[-1])
            loop=True
            while(loop==True):
                each_line=file_read.readline()
                if(each_line.find('[Term]')>=0):
                    count+=1
                    each_hpo_umls.append('NA')
                    print(each_hpo_umls," " ,count)
                    hpo_umls_df.append(each_hpo_umls)
                    each_hpo_umls=[]
                    each_line=file_read.readline()
                    if(each_line.find('id: HP:')>=0):
                        each_hpo_umls.append(each_line.strip().split(':')[-1])
                        continue
                if(each_line.find('xref: UMLS:')>=0):
                    loop=False
                    each_hpo_umls.append(each_line.strip().split(':')[-1])
            hpo_umls_df.append(each_hpo_umls)
            print(each_hpo_umls," " ,count)
hpo_umls_df=pd.DataFrame(hpo_umls_df,columns=['HPO','UMLS'])
hpo_umls_df.to_csv(r'hpo_umls_map.csv', index = False)               
```
